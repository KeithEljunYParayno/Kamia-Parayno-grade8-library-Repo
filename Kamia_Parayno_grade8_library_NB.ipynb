```python
import requests
import json

url = "https://raw.githubusercontent.com/KeithEljunYParayno/Kamia-Parayno-grade8-library-Repo/refs/heads/main/books.json"

response = requests.get(url)
data = response.json()

print(data)
```

```python
import firebase_admin
from firebase_admin import credentials, db

cred = credentials.Certificate("firebase_key.json")
firebase_admin.initialize_app(cred, {
    'databaseURL': 'https://kamia-parayno-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/'
})
```

```python
ref = db.reference("library")
ref.set(data)

print("Data uploaded successfully!")
```

```python
ref = db.reference("library/books")
books = ref.get()

print(books)
```

```python
for book in books:
    print(book["title"])
```

```python
for book in books:
    print(book["title"], "-", book["details"]["available"])
```

```python
total = 0

for book in books:
    total += book["details"]["copies"]

print("Total number of books:", total)
```
